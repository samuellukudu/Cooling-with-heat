# Simulation tools

## I. Computational fluid dynamics for general heat transfer and flow simulations

- OpenFoam (Windows, Linux and Mac(less well tested than Linux))

- SU2 (Linux, Mac)

- SimWorks (windows and linux)

## II. Thermal and multi-physics simulations

- Energy2D (Windows, Linux, Mac, Raspberry)

- OpenModelica (windows, linux, mac, docker)

- Thermo (via NanoHUB)

## III. Finite Element Analysis (FEA) for structural and thermal analysis 

- Z88 (windows, linux, macos)

- Elmer (linux and windows)

### Other tools

- Calculix ()

- FreeCAD (FEM workbench)

### Recommendations

* Start with Energy2D and SimWorks or OpenFOAM for initial explorations and gaining insights into heat transfer in fabrics.

* If you need more advanced or coupled simulations, explore OpenModelica, CalculiX, or consider using a combination of tools like FreeCAD with an FEA solver.
For specific thermoelectric device optimization, use the Thermo simulator available through nanoHUB.
