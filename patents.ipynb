# Patents

## I. Passive cooling mechanisms

### Evaporative cooling textiles:

* One patent focuses on chemical-free cooling fabrics that manage moisture transfer and evaporation for reduced surface temperature.

* Another patent application describes an evaporative cooling clothing system that includes a clothing article, a liquid reservoir, and a hollow transport system to expose the liquid for evaporation.

* A patent exists for an evaporative cooling towel with a method of activation.

### Cooling fabric and materials:

* A patent has been granted for a cooling fabric.

* Another patent describes a fabric cloth inspired by Saharan silver ants that can replace air conditioners, according to SRM University AP.

* A patent application discusses cooling fabrics made from materials like n-Octacosane.

### Passive radiative cooling textiles:

* A patent exists for a spectrally selective textile for passive radiative outdoor personal cooling, which exhibits specific properties for infrared radiation transmittance and reflectivity.

## II. Active cooling mechanisms

### Thermoelectric cooling (TEC) in clothing:

* Several patents discuss incorporating TECs in clothing for both heating and cooling, often using Peltier devices.

* One patent describes a heating/cooling garment that circulates a heating medium heated and cooled by a Peltier device.

* Other patents describe direct placement of Peltier devices on the carotid body area or the neck for localized cooling or heating.

### Liquid cooling garments (LCGs):

* Patents exist for liquid loop garments for heating and cooling individuals, including closed systems circulating liquid.

* Another patent describes a body cooling apparatus using a tubing system to circulate fluid and remove heat.

* A cooling garment with tubes for circulating a liquid water-ammonia solution for evaporative cooling is also patented.

## III. Phase change material (PCM) integration

### PCM applications in garments:

* NASA's initial work on PCMs for space suits and Gateway Technologies' acquisition of exclusive patents to make PCMs accessible to consumers in apparel are important historical developments.

* A patent exists for phase change material compositions and methods for their use to lower surface friction and wear, which could be relevant to clothing applications.

* Other patents focus on aerogel compositions with thermal storage capacity incorporating PCM coatings, particle mixtures, or PCM materials confined within the porous network of an aerogel composition.

# Wearables patents

## I. Wearable thermoelectric generators (TEGs)

**US20130087180A1 - Wearable thermoelectric generator system**: This patent describes a system that includes a thermoelectric generator, a heat collector in contact with the skin, and a heat exchanger exposed to ambient air.

**Fabric-based TEGs**: Research explores incorporating thermoelectric materials into fabrics, including textile-based TEGs with carbon fibers and flexible TEGs made with Mg-based materials.

**Flexible and Stretchable TEGs**: One invention focuses on mechanically flexible and stretchable thermoelectric devices for both thermoregulation (cooling and heating) and power generation, without needing a rigid heat sink.

**Wearable TEGs for Biomedical Applications**: A flexible thermoelectric wearable architecture has been developed for continuous monitoring of physiological signals, powering wireless bioelectronics by harvesting body heat.

**US20160049569A1 - Thermoelectric power source for personal electronics and wearable items**: This patent describes a wearable power source with a thermoelectric generator (e.g., Peltier tile) that uses body heat to generate electricity for personal electronics. 

## II. Wearable thermoregulation devices

**US12096828 - Wearable thermoregulation device, system and method responsive to hot flashes**: This patent describes a wearable device using a thermoelectric device to heat or cool a control surface on the user's skin.

**US8397518B1 and WO2013124866A2 - Apparel with integral heating and cooling device**: These patents detail clothing incorporating a thermoelectric module, heat sink, and wetting material for cooling through evaporation.

**WO2004014169A2 - Temperature regulated clothing**: This patent describes clothing with a battery-powered thermoelectric module for heating or cooling, potentially with an integrated circuit for temperature control. 

## III. Thermogalvanic cells in wearables

**Thermogalvanic Hydrogels**: Research shows promise for thermogalvanic hydrogels for self-powered on-body temperature and strain sensing in wearable electronics.

**Gel Electrolyte-Based Thermogalvanic Generator**: A flexible, wearable device has been developed using a gel electrolyte-based thermogalvanic generator to harvest body heat for self-powered medical electronics and potentially even cooling fever patients.

**US2882329A - Thermogalvanic cell**: This patent describes a thermogalvanic cell with a cation-permeable ion-exchange resin membrane sandwiched between two metal electrodes.

**Ultrastrong, Flexible Thermogalvanic Armor**: An advanced design features high efficiency, power density, stretchability, strength, and toughness, overcoming the trade-off between thermoelectric and mechanical properties in conventional thermocells. 

## IV. Energy harvesting clothing

**Solar Energy Harvesting Fabric**: Research has demonstrated fabrics with integrated miniature solar cells for powering wearable and mobile devices.

**Wearable Power Harvesting System**: A system using a knitted fabric rectenna can harvest radio-frequency energy (e.g., WLAN frequencies) to power wearable electronics.

**Energy Harvesting from Clothing with Triboelectric Nanogenerators**: A system has been developed for generating electricity by simply sliding a zein film assembly on the surface of clothes, using zein/cloth direct current triboelectric nanogenerators (TENGs).

**Energy Harvesting from Human Movements**: Research is ongoing to develop wearable electronics with integrated systems to convert the mechanical energy of human movements into electrical energy.
